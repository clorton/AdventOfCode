```python
def process(a):
    b = a%8
    b ^= 2
    c = a//2**b
    b ^= 3
    b ^= c
    print(f"{a} -> {b%8} ({a}, {b}, {c})")
```

```python
for a in range(8):
    process(a)
```

```python
def process2(a):
    while True:
        b = a%8
        b ^= 2
        c = a//2**b
        b ^= 3
        b ^= c
        print(f"{a} -> {b%8} ({a}, {b}, {c})")
        a //= 8
        if a == 0:
            break
```

```python
for a in range(8,15):
    process2(a)
```

```python
def process3(a, goal):
    b = a%8
    b ^= 2
    c = a//2**b
    b ^= 3
    b ^= c
    if b%8 == goal:
        print(f"{a} -> {b%8} ({a}, {b}, {c})")
    return b%8
```

```python
program = [0,3,3,0,5,5,3,4,3,1,5,7,2,1,4,2]
value = 0
for i, output in enumerate(program):
    print(f"Looking for program[{i}] = {output}...")
    for a in range(value*8, (value+1)*8):
        if process3(a, output) == output:
            print(f"Found {a} for {output}")
            value = a
            break
```

```python
program = [0,3,3,0,5,5,3,4,3,1,5,7,2,1,4,2]
values = [[0]]
for i, output in enumerate(program):
    print(f"Looking for program[{i}] = {output}...")
    new_values = []
    for value in values[-1]:
        for a in range(value*8, (value+1)*8):
            if process3(a, output) == output:
                print(f"Found {a} for {output}")
                new_values.append(a)
    values.append(new_values)
```

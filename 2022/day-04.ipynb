```python
from pathlib import Path
INPUT_FILE = Path.cwd() / "day-04.txt"
```

```python
with INPUT_FILE.open("r") as file:
    input = [line.strip() for line in file.readlines()]
```

```python
input[0:8]
```

```python
count = 0
for ranges in input:
    range1, range2 = ranges.split(",")
    start, end = list(map(int, range1.split("-")))
    range1 = set(range(start, end+1))
    start, end = list(map(int, range2.split("-")))
    range2 = set(range(start, end+1))
    intersection = range1 & range2
    if (range1 == intersection) or (range2 == intersection):
        count += 1
        print(f"Found contained range in {ranges}")

print(f"Part1: Count = {count}")
```

```python
count = 0
for ranges in input:
    range1, range2 = ranges.split(",")
    start, end = list(map(int, range1.split("-")))
    range1 = set(range(start, end+1))
    start, end = list(map(int, range2.split("-")))
    range2 = set(range(start, end+1))
    intersection = range1 & range2
    if len(intersection) > 0:
        count += 1

print(f"Part2: Count = {count}")
```
